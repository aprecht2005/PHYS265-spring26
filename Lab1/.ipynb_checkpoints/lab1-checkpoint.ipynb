## Part 1: The Gravitational Potential of the Earth
----

The gravitational potential at a distance $r$ from a mass $M$ is given by:

$$\Phi(r) = -\frac{GM}{r}$$

where G is the gravitational constant.

1. Code up a function that takes as an input the mass of a body $M$ , the cartesian location of the
mass $(x_M , y_M)$, and the location of some distant evaluation point $(x, y)$. Have the function
return the potential at the evaluation point. You may consider approximating the Earth as
a point particle, but in which case, be careful! For in that approximation, the potential
contains a singularity. In our function, use all modern best practices for coding.
E.g. type hints, type checks, pre and post conditions, doc strings, etc.

```python
def GravPotential(M, x_m, y_m, x, y): 
    """A function that calculates the gravitational potential at a distance from a mass

    Arguments: 
        M (float): the mass
        x_m (float): the x-coordinate of the mass
        y_m (float): the y-coordinate of the mass 
        x (float): the x-coordinate of the distance from the mass
        y (float): the y-coordinate of the distance from the mass

    Returns:
        g_potential (float): the gravitational potential at a distance r from the mass
    """
    try:
        if not (isinstance(M, float) and isinstance(x_m, float) and isinstance(y_m, float) and isinstance(x, float) and isinstance(y, float)):
            raise TypeError()
    except TypeError: 
        return "All inputs must be floats" 
    else:
        x_dist = x - x_m 
        y_dist = y - y_m
        r = np.sqrt(x_dist**2 + y_dist**2)
        if np.isclose(r, 0): 
            return np.nan
        g_potential = -(G_const * M) / r 
        assert isinstance(g_potential, float)
        return g_potential 
```

2. Make a 1D plot of the absolute value of the gravitational potential $|\Phi|$ as a function of distance $x$ from the surface of the Earth, assuming $y = 0$.
Plot out to $1.5\times$ the distance from the Earth to the Moon. Because the potential falls quickly with r, make the y-axis logarithmic to visualize it better.

```python
x_vals = np.linspace(-dist_em, dist_em, 1000) 
GravPotential_v = np.vectorize(GravPotential)
y_vals = GravPotential_v(Mass_e, 0.0, 0.0, x_vals, 0.0)
abs_y_vals = np.abs(y_vals)
fig, ax = plt.subplots() 
ax.plot(x_vals, abs_y_vals)
ax.set_yscale('log')
ax.set_xlabel('x')
ax.set_ylabel('Gravitational Potential')
fig.tight_layout()
```

## Part 3: The Gravitational Force Field of the Earth-Moon System
----
The gravitational force $\vec{F}$ 
that a mass $M_1$ exerts on a mass $m_2$ is:

$$\vec{F}_{21} = -G \frac{M_1 m_2}{|\vec{r}_{21}|^2} \hat{r}_{21}$$

where $\vec{r}_{21}$ is the displacement vector from $M_1$ to $m_2$.

1. Code up a function that takes as an input the mass of a body $M_1$, 
the mass of a secondary body $m_2$,
the cartesian location of the first mass $(x_1, y_1)$, 
and the cartesian location of the second body $(x_2, y_2)$. 
Have the function return the two gravitational force components, 
$F_x$ and $F_y$, experienced by the mass $m_2$ 
from the force exerted by $M_1$.
Note that the gravitational force contains a singularity, 
and so if you made the approximation that the Earth is a finite point,
you will need to avoid it.
Like for the 1D potential, make sure it follows modern best-coding practices.

```python
def GravPotential2(M_1, M_2, x_1, y_1, x_2, y_2):
    """A function that calculates the gravitational potential that one mass exerts on another

    Arguments: 
        M_1 (float): the mass of the first object
        M_2 (float): the mass of the second object
        x_1 (float): the x-coordinate of the first mass
        y_1 (float): the y-coordinate of the first mass 
        x_2 (float): the x-coordinate of the second mass
        y_2 (float): the y-coordinate of the second mass

    Returns:
        F_x (float): the gravitational force in the x-direction that mass 1 exerts on mass 2
        F_y (float): the gravitational force in the y-direction that mass 1 exerts on mass 2
    """
    try:
        if not (isinstance(M_1, float) and isinstance(M_2, float) and isinstance(x_1, float) and isinstance(y_1, float) and isinstance(x_2, float) and isinstance(y_2, float)):
            raise TypeError()
    except TypeError: 
        return "All inputs must be floats" 
    else:
        r_x_dist = x_2 - x_1
        r_y_dist = y_2 - y_1
        r_squared = r_x_dist**2 + r_y_dist**2
        r = np.sqrt(r_squared)
        if np.isclose(r, 0): 
            F_x = np.nan
            F_y = np.nan
        else:
            F_x = -(G_const * M_1 * M_2 * r_x_dist) / (r_squared * r)
            F_y = -(G_const * M_1 * M_2 * r_y_dist) / (r_squared * r)
            assert isinstance(F_x, float)
            assert isinstance(F_y, float)
    return F_x, F_y
```

2. Assuming the Earth is still at the origin and the Moon is still at 
$(x_M, y_M) = \big( \frac{d_{EM}}{\sqrt{2}}, \frac{d_{EM}}{\sqrt{2}} \big)$, 
make a 2D streamplot of the gravitational force the Earth-Moon system 
will exert on the Apollo 11 command module.
Plot the force over the same x-y range of $-1.5 d_{EM}$ to $+1.5 d_{EM}$.
Add a nice color bar, which will again probably need to be logarithmic.
Mark the location of the earth with a black circle, and the moon with a black square.
Add a legend.

```python
ny, nx = 500, 500
x = np.linspace(-1.5 * dist_em, 1.5 * dist_em, nx)
y = np.linspace(-1.5 * dist_em, 1.5 * dist_em, ny)
x_vals, y_vals = np.meshgrid(x, y)

Force_x = np.zeros((ny, nx)) 
Force_y = np.zeros((ny, nx))

for y in range(ny): 
    for x in range(nx): 
        force_ea_x, force_ea_y = GravPotential2(Mass_e, Mass_a, 0.0, 0.0, x_vals[y, x], y_vals[y, x])
        force_ma_x, force_ma_y = GravPotential2(Mass_m, Mass_a, dist_em / np.sqrt(2), dist_em / np.sqrt(2), x_vals[y, x], y_vals[y, x])
        Force_x[y, x] = (force_ea_x + force_ma_x)
        Force_y[y, x] = (force_ea_y + force_ma_y)
#Summing the forces from the Earth and Moon
fig, ax = plt.subplots()
mag = np.hypot(Force_x, Force_y)
colors = np.log(mag)
stream = ax.streamplot(x_vals, y_vals, Force_x, Force_y, color = colors, cmap = cm.magma, linewidth = 1, density = 1.5)
fig.colorbar(stream.lines, label = 'Force Magnitude')
ax.set_aspect('equal')
ax.plot(0.0, 0.0, 'ko', ms = 10, label = 'Earth')
ax.plot(dist_em / np.sqrt(2), dist_em / np.sqrt(2), 'ks', ms = 10, label = 'Moon')
fig.legend()
```

3. You will note there is a point between the Earth and the Moon where the field lines seem to vanish. This is the point where the Earth's and Moon's gravitational forces cancel eachother. By setting $|\vec{F}_{moon}| = |\vec{F}_{earth}|$, calculate the radial distance of the "balance point." From there, identify the $x$ and $y$ coordinates of the balance point. Typeset your derivation (show multiple steps!) in latex in a markdown cell below. Then, calculate it numerically, and draw the answer onto a new version of your force field plot. Again mark the location of the earth with a black circle, and the moon with a black square; mark the location of the balance point with a red x. Add a legend.

**Your Derivation Goes in the markdown cell below:**

And your updated plot, including the balance point, goes in the cell below.

First we need to find the magnitude of the Moon and the Earth's forces at their distance from the balance point. From this we get that the Earth's magnitude is $\frac{G M_e}{r^2}$ and the Moon's is $\frac{G M_m}{(d - r)^2}$. $G$ represents the gravitational constant, $M_e$ represents the mass of the Earth, $M_m$ represents the mass of the Moon, $r$ represents the radial distance from the Earth and Moon to the balance point, and $d$ represents the distance between the Earth and Moon. If we set them equal to each other and square root both sides to simplify calculations we get $\frac{\sqrt{M_e}}{r} = \frac{\sqrt{M_m}}{d - r}$. Multiply both sides by the denominators and distribute to get $\sqrt{M_e}d - \sqrt{M_e}r = \sqrt{M_m}r$. Move the r term to other side and undistribute to get $r(\sqrt{M_m} - \sqrt{M_e}) = \sqrt{M_e}d$. Divide both sides to isolate r and our final expression for r becomes $r = \frac{\sqrt{M_e}d}{\sqrt{M_m} + \sqrt{M_e}}$. Since we know the Moon is at a distance $\frac{d}{\sqrt{2}}$ from the Earth we can just substitute in r in place of d to find the location of the balance point since it must be between the Earth and Moon. Thus our x and y coordinates are $x = \frac{r}{\sqrt{2}}$ and $y = \frac{r}{\sqrt{2}}$

```python
dist_bal = dist_em * (np.sqrt(Mass_e) / (np.sqrt(Mass_e) + np.sqrt(Mass_m)))
x_bal = dist_bal / np.sqrt(2)
y_bal = dist_bal / np.sqrt(2)

ny, nx = 500, 500
x = np.linspace(-1.5 * dist_em, 1.5 * dist_em, nx)
y = np.linspace(-1.5 * dist_em, 1.5 * dist_em, ny)
x_vals, y_vals = np.meshgrid(x, y)

Force_x = np.zeros((ny, nx)) 
Force_y = np.zeros((ny, nx))

for y in range(ny): 
    for x in range(nx): 
        force_ea_x, force_ea_y = GravPotential2(Mass_e, Mass_a, 0.0, 0.0, x_vals[y, x], y_vals[y, x])
        force_ma_x, force_ma_y = GravPotential2(Mass_m, Mass_a, dist_em / np.sqrt(2), dist_em / np.sqrt(2), x_vals[y, x], y_vals[y, x])
        Force_x[y, x] = (force_ea_x + force_ma_x)
        Force_y[y, x] = (force_ea_y + force_ma_y)

fig, ax = plt.subplots(figsize = (7, 7))
mag = np.hypot(Force_x, Force_y)
colors = np.log(mag)
stream = ax.streamplot(x_vals, y_vals, Force_x, Force_y, color = colors, cmap = cm.magma, linewidth = 1, density = 1.5)
fig.colorbar(stream.lines, label = 'Force Magnitude')
ax.set_aspect('equal')
ax.plot(0.0, 0.0, 'ko', ms = 10, label = 'Earth')
ax.plot(dist_em / np.sqrt(2), dist_em / np.sqrt(2), 'ks', ms = 10, label = 'Moon')
ax.plot(x_bal, y_bal, 'rx', ms = 10, label='Balance Point')
fig.legend(loc = 'upper left')
fig.tight_layout()
```

## Part 4: Altitude of the Saturn V Rocket
----
Rockets function by conservation of momentum.
The ejection of fuel backwards propels the rocket forwards.
The change in a rocket's velocity $\Delta v$ as a 
function of time $t$ (under a set of simplifying assumptions) 
is given by the Tsiolkovsky rocket equation:

$$\Delta v(t) = v_e \ln \left(\frac{m_0}{m(t)}\right) - gt$$

where $m_0$ is the initial "wet" mass (fuel + rocket parts + payload), 
$m(t) = m_0 - \dot{m}t$ is the mass at time $t$, 
$\dot{m}$ is the fuel burn rate (which we will assume to be constant), 
$v_e$ is the fuel exhaust velocity, and $g$ is the gravitational acceleration.

You can find the altitude $h$ of the rocket at "burnout" 
(when all the fuel is used up) by integrating the velocity over time:

$$h = \int_{0}^{T} \Delta v(t) \, dt$$

where $T$ is the total burn time of the rocket, 
which by conservation of momentum arguments is:

$$T = \frac{m_0 - m_f}{\dot{m}}$$

where $m_f$ is the final "dry mass" of the rocket, 
once all the fuel is burned.

1. Calculate the burn time $T$ for the first stage of
the new Saturn V rocket that NASA is planning
to use to carry the Apollo capsule.

```python
burn_time = (wet_mass - dry_mass) / burn_rate
print(burn_time)
```

2. Define a function that returns the change in velocity $\Delta v(t)$.
It should take all necessary arguments, meaning: the time, 
wet mass, dry mass, burn rate, exhaust speed, and gravitational constant $g$. 
Have it return the change in velocity.
Note that the rocket eventually runs out of fuel, 
so you should design your function carefully so that 
$\Delta v$ eventually becomes zero after all fuel is spent.
Again, document and guard your function according to best practices.

```python
def RocketVelocity(t, m0, mf, mdot, ve, g):
    """ A function that calculates the change in a rockets velocity at any time t 

    Arguments: 
        t (float): the time of the change in velocity
        m0 (float): the initial wet mass of the rocket
        mf (float): the final dry mass of the rocket after the fuel has been burned
        mdot (float): the burn rate of the rocket's fuel
        ve (float): the fuel exhaust velocity of the rocket
        g (float): the gravitational acceleration of earth

    Returns: 
        delta_v (float): the change in velocity of the rocket at a time t
    """
    try:
        if not (isinstance(t, float) and isinstance(m0, float) and isinstance(mf, float) and isinstance(mdot, float) and isinstance(ve, float) and isinstance(g, float)):
            raise TypeError()
    except TypeError: 
        return "All inputs must be floats"
    else:
        mt = m0 - (mdot * t)
        if mt - mf <= 0: 
            return 0.0
        else:
            delta_v = ve * np.log(m0 / (m0 - (mdot * t))) - g * t 
            assert isinstance(delta_v, float)
            return delta_v
```

3. Compute the altitude of the rocket at the end of the burn.
Do this by numerically integrating from $t=0$ (launch) to $t=T$ (burnout).
Use the `scipy.integrate.quad()` function.

```python
altitude_r = sp.integrate.quad(RocketVelocity, 0, burn_time, args = (wet_mass, dry_mass, burn_rate, saturn_v, g_earth))
print(altitude_r)
```
